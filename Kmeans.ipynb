```python
import pandas as pd
```

```python
data = iris.data
data
```

```python
df = pd.DataFrame(data,columns=['Sepallength','SepalWidth','PetalLength','PetalWidth'])
```

```python
df.head(5)
```

```python
df= df.drop(['Sepallength','SepalWidth'],axis=1)
```

```python
from sklearn.cluster import KMeans
from sklearn.preprocessing import MinMaxScaler
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
plt.scatter(df.PetalLength,df['PetalWidth'])
plt.xlabel('Length')
plt.ylabel('Width')
plt.show()
```

```python
km = KMeans(n_clusters = 2)
```

```python
y_predicted = km.fit_predict(df[['PetalLength','PetalWidth']])
y_predicted
```

```python
df['cluster'] = y_predicted
df.head(5)
```

```python
km.cluster_centers_
```

```python
df1 = df[df.cluster==0]
df2 = df[df.cluster==1]
```

```python
plt.scatter(df1.PetalLength,df1['PetalWidth'],color='green')
plt.scatter(df2.PetalLength,df2['PetalWidth'],color='red')
plt.xlabel('Length')
plt.ylabel('Width')
plt.show()
```

```python
scalar = MinMaxScaler()
```

```python
scalar.fit(df[['PetalWidth']])
```

```python
scalar.fit(df[['PetalLength']])
```

```python
df['PetalWidth'] = scalar.transform(df[['PetalWidth']])
```

```python
df['PetalLength'] = scalar.transform(df[['PetalLength']])
```

```python
df.head(5)
```

```python
km = KMeans(n_clusters = 2)
y_predicted = km.fit_predict(df[['PetalLength','PetalWidth']])
y_predicted
```

```python
km.cluster_centers_
```

```python
df1 = df[df.cluster==0]
df2 = df[df.cluster==1]
```

```python
plt.scatter(df1.PetalLength,df1['PetalWidth'],color='green')
plt.scatter(df2.PetalLength,df2['PetalWidth'],color='red')
plt.xlabel('Length')
plt.ylabel('Width')
plt.show()
```

```python
sse = []
```

```python
k_rng = range(1,10)
```

```python
for k in k_rng:
 km = KMeans(n_clusters = k)
 km.fit(df[['PetalLength','PetalWidth']])
 sse.append(km.inertia_)
```

```python
sse
```

```python
plt.plot(k_rng,sse)
plt.xlabel(k)
plt.ylabel('Sum of squared errors')
plt.show()
```
